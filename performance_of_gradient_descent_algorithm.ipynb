```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])
```

```python
m = 0
c = 0
learning_rate = 0.01
iterations = 50

n = len(X)
cost_history = []
```

```python
for i in range(iterations):

    y_pred = m * X + c

    # Calculate cost (MSE)
    cost = (1/n) * sum((y - y_pred)**2)
    cost_history.append(cost)

    # Compute gradients
    dm = (-2/n) * sum(X * (y - y_pred))
    dc = (-2/n) * sum(y - y_pred)

    # Update parameters
    m = m - learning_rate * dm
    c = c - learning_rate * dc
```

```python
print("Final slope (m):", m)
print("Final intercept (c):", c)
print("Final Cost:", cost_history[-1])
```

```python
#Plot regression line
plt.scatter(X, y)
plt.plot(X, m*X + c)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Gradient Descent Line Fit")
plt.show()
```

```python
#Plot cost reduction
plt.plot(cost_history)
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.title("Cost Reduction During Gradient Descent")
plt.show()
```
